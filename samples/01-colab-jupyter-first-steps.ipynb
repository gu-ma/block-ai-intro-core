## 3) Mini random idea generator

```python
import random

subjects = ["city", "forest", "face", "machine", "archive"]
styles = ["minimal", "chaotic", "nostalgic", "futurist", "organic"]
constraints = ["2 colors only", "no text", "vertical format", "one geometric shape", "high contrast"]

prompt = f"Create a {random.choice(styles)} composition about {random.choice(subjects)} using {random.choice(constraints)}."
print(prompt)
```
